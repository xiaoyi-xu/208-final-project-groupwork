### 2.3 Data description
In this project, the analysis uses data of the heart attack from [Kaggle](https://www.kaggle.com/rashikrahmanpritom/heart-attack-analysis-prediction-dataset).
The original data set contains 303 sets of observations and 14 different variables. Each observation represents various information of each patient. 
*  dependent variable
** output: Categorical variable used to indicate whether it is prone to heart attack (0 = less chance of heart attack, 1 = more chance of heart attack).
* categorical independent variables
** sex: Gender of the patient (0 = male, 1 = female).
** cp: Chest Pain type (1 = typical angina, 2 = atypical angina, 3 = non-anginal pain, 4 = asymptomatic).
** fbs: Whether fasting blood sugar is greater than 120 mg/dl (0 = false, 1 = true).
** restecg: Resting electrocardiographic results (0 = normal, 1 = having ST-T wave abnormality (T wave inversions and/or ST elevation or depression of > 0.05 mV), 2 = showing probable or definite left ventricular hypertrophy by Estes' criteria).
** exng: Whether exercise induce angina (0 = no, 1 = yes).
** slp: Slope (0-2).
** caa: Number of major vessels (0-3).
** thall: Thalium Stress Test result (0-3).
* numerical independent variables
** age: Age of the patient.
** trtbps: Resting blood pressure (in mm Hg).
** chol: Cholestoral in mg/dl fetched via BMI sensor.
** thalachh: Maximum heart rate achieved.
** oldpeak: Previous peak.

## 4. Inferential analysis

### 4.1 Variable selection

* simple linear regression

In order to better understand this set of data and select variables for further classification, we first simply fit the linear regression model to the output to see which variables are important. 

```python
%run ../code/4.1.1.py
```

It can be seen from the R output that  the p-value for T-test of most variables is far less than 0.05, that means most variables are statistically significant, only age, trtbps, chol, fbs, restecg are not significant. And for the F-test of the entire model, the p-value is less than 1.48e-85, the model is significant. This result seems acceptable. The adjusted R-squared is 0.764, which shows that the explanatory ability of this model need further improvement. After that, we made diagnostic plots of the model. From â€œresiduals vs fittedâ€ plot, we can see that residuals have obvious patterns, which also shows that it is not appropriate to use simple linear regression model for fitting. 

* logistic regression with L1 penalty

Since we see that residuals have a clear grouping trend in the diagnostic plot of the linear regression model, we think that we should introduce a logistic regression model to improve.Also, we add a regularization term to the logistic equation in order to prevent overfitting of the model. Therefore, we use logistic regression with L1 penalty and use logistic regression with L2 penalty with C(regularization parameter) between 0.05 and 1 and then compare the model accuracy.

```python
%run ../code/4.1.2.py
```

In the plot, there is not much difference between the accuracy results of logistic regression with L1 penalty and use logistic regression with L2 penalty. When C=0.01, the accuracy of L1 test and L2 test achieve the toppest together, and the accuracy is 0.83516. Therefore, we choose C=0.1 and use logistic regression with L1 penalty to select variables. The final variables we choose are age, sex, cp, trtbps, chol, thalachh, oldpeak, caa and thall.

```python
n_est = {}
for i in range(10,100):
    for k in range(1,25):
        rf = RandomForestClassifier(n_estimators=i, max_depth=k, criterion='entropy', random_state=123)
        rf.fit(X_tr, y_tr)
        accuracy = rf.score(X_te, y_te)
        n_est[i, k] = round(accuracy*100,2)
```

```python
print(f'(n_estimator, max_depth) at max accuracy: {max(n_est , key = n_est.get)}')
```

Thus, we can construct random forest model with n_estimator = 21 and max_depth = 1.

```python
rf = RandomForestClassifier(n_estimators=21, max_depth=1, criterion='entropy', random_state=123)

model(rf,X_tr,y_tr,X_te,y_te)
```

### 4.3 Cross Validation

In the last part, we use the normal method to split the data into training data and test data. Now we can use K fold cross validation to check which model would stay steady. We will choose K = 5.

```python
%run ../code/4.2.4.py
X = pd.concat([X_tr,X_te])
y = pd.concat([y_tr,y_te])
accuracy = []
```

```python
accuracy.append(kfold(svc,X,y))
accuracy.append(kfold(svc2,X,y))
accuracy.append(kfold(rf,X,y))
accuracy.append(kfold(xgboost,X,y))
accuracy.append(kfold(knn,X,y))
```

```python
model_score['5-Fold score'] = accuracy
model_score
```

We can see that the accuracy score of these models decrease when we use K fold cross validation. And the SVC model with sigmoid kernel is the most stable model, whose accuracy score decreases minimally. Comparing with two different validation methods, SVC with linear kernel is still the model with the highest accuracy score for two methods and SVC with sigmoid kernel is still the model with the lowest accuracy score for two methods.
